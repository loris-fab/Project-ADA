## Handling data



- # Part 1 : Data models

#### **Modle Plat (Flat Model)**
- **Description :** Un seul type d'entit, mmes attributs. 
- **Exemples :** CSV, fichiers de log.
- **Avantages :** Simple, facile  manipuler.
- **Inconvnients :** Limit pour les structures complexes.

#### **Modle Relationnel (Relational Model)**
- **Description :** Plusieurs entits connectes par des relations. Donnes en tables (colonnes et lignes).
- **Exemples :** MySQL, PostgreSQL, Oracle.
- **Avantages :** Flexible, bon pour les donnes complexes, support SQL.
- **Inconvnients :** Complexe pour des structures trs imbriques.

#### **Modle Document (Document Model)**
- **Description :** Hirarchie d'entits, donnes sous forme de documents (JSON, XML).
- **Exemples :** MongoDB, fichiers JSON/XML.
- **Avantages :** Flexible, bon pour les donnes non structures.
- **Inconvnients :** Moins efficace pour les jointures, redondance possible.

#### **Modle Rseau (Network Model)**
- **Description :** Rseau complexe d'entits, modle de graphe.
- **Exemples :** Neo4j, RDF.
- **Avantages :** Excellente pour les donnes hautement interconnectes.
- **Inconvnients :** Complexe  modliser et interroger.

#### **Choix du Modle**
- Dpend du type de donnes, complexit et oprations. 
- Modle rseau pour interconnexions, document pour hirarchie, relationnel pour structure classique.
  
#### **SQL vs Pandas**
- **SQL :** Utilis pour interroger et grer les donnes dans des modles relationnels avec des commandes comme SELECT, UPDATE, DELETE, et JOIN.
- **Pandas (bibliothque Python) :** Similaire  SQL mais avec des capacits supplmentaires de programmation fonctionnelle. Pandas est rapide, sintgre bien avec Python et les bibliothques de visualisation, mais ncessite que toutes les donnes tiennent en mmoire.

- # Part 3 : Data Wrangling (Nettoyage des Donnes)
- **Dfinition :** Prparation des donnes brutes pour l'analyse : nettoyage, transformation, standardisation.
- **Problmes courants :** Donnes manquantes, incorrectes, incohrences, doublons.
- **Importance :** Reprsente 50-80 % du temps en data science.

### **Types de Problmes de Donnes**
- **Donnes Manquantes :** Valeurs absentes, traitement avec interpolation ou suppression.
- **Donnes Incorrectes :** Erreurs de mesure ou de traitement.
- **Incohrences :** Reprsentations diffrentes du mme type de donnes.

### **Stratgies de Nettoyage**
- **Automatisation + Visualisation Interactive :** Aide  dtecter et corriger les anomalies.
- **Visualisation pour Diagnostic :** Identifie les problmes (ex : outliers, valeurs nulles).

### **Exemples de Problmes de Donnes**
- **Histoires dhorreur :** Lettres incorrectes, erreurs dattribution (ex : hommes  enceintes ).
- **Graphes et Matrices :** Utilisation pour visualiser les connexions et incohrences.

### **Conseils Pratiques**
- **Avant Analyse :** Vrifier les donnes manquantes, corrompues, et les transformer au bon format.
- **Documentation et Provenance :** Toujours chercher les sources et la documentation des donnes.
- **Ne pas Sur-nettoyer :** viter de perdre des informations importantes en corrigeant trop.
